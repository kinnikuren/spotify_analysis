```python
import pandas as pd
import glob
import json
```

```python
streaming_history = pd.read_json("StreamingHistory3.json")

print(streaming_history.head())
```

```python
streaming_history_files = glob.glob('StreamingHistory*')
df_list = []
for file in streaming_history_files:
    df = pd.read_json(file)
    print(file)
    print(len(df))
    df_list.append(df)
    
streaming_history = pd.concat(df_list)
print(len(streaming_history))
print(streaming_history.head())
print(streaming_history.tail())
```

```python
grouped_by_track = streaming_history.groupby(['artistName', 'trackName'])['endTime'].count().reset_index()

grouped_by_track = grouped_by_track.rename(columns={'endTime': 'count'}).sort_values('count', ascending=False)

print(grouped_by_track.head(50))
```

```python
#library = pd.read_json('YourLibrary.json')
with open('YourLibrary.json', encoding='utf8') as f:
    data = json.load(f)
    
print(data.keys())

library = pd.DataFrame.from_dict(data['tracks'])

print(library.head())
print(len(library))
```

```python
library_by_artist = library.groupby('artist')['track'].count().reset_index().rename(columns={'track': 'count'}).sort_values('count', ascending=False)

print(library_by_artist.head(50))
```
